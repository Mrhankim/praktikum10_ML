```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
data = {
    'Actual': [
        'Lulus','Lulus','Lulus','Lulus','Lulus',
        'Tidak Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus'
    ],
    'Predicted': [
        'Lulus','Lulus','Lulus','Tidak Lulus','Tidak Lulus',
        'Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus'
    ]
}

df = pd.DataFrame(data)
```

```python
# Hitung confusion matrix
labels = ['Lulus', 'Tidak Lulus']
cm = confusion_matrix(df['Actual'], df['Predicted'], labels=labels)
```

```python
print("Confusion Matrix")
print(pd.DataFrame(cm, columns=['Pred Lulus','Pred Tidak Lulus'],
                        index=['Actual Lulus','Actual Tidak Lulus']))
```

```python
# Hitung metrik
accuracy = accuracy_score(df['Actual'], df['Predicted'])
precision = precision_score(df['Actual'], df['Predicted'], pos_label='Lulus')
recall = recall_score(df['Actual'], df['Predicted'], pos_label='Lulus')

print("\nAccuracy  :", accuracy)
print("Precision :", precision)
print("Recall    :", recall)
```
